```python
import matplotlib.pyplot as plt
import numpy as np

# Function to create subplots for each emotion type
def create_subplots(ax, data, title, ylabel):
    for i, emotion_data in enumerate(data):
        for j in range(len(emotion_data)):

            ax[j, i].plot(emotion_data[j])
            ax[j, i].set_title(f"{title} - {emotions[j]} - {info_matrix[j, i]}")
            ax[j, i].set_xlabel("Epoch")
            ax[j, i].set_ylabel(ylabel)
            ax[j, i].grid(True)

# List of emotions
emotions = ['ang', 'hap', 'sad']

# Information matrix
info_matrix = np.array([
    ['T1_rep', 'T2_rep', 'T1_zero', 'T2_zero'],
    ['T1_rep', 'T2_rep', 'T1_zero', 'T2_zero'],
    ['T1_rep', 'T2_rep', 'T1_zero', 'T2_zero']
])

# Data for validation loss
val_data = [
    [val_T1_rep_ang, val_T1_rep_hap, val_T1_rep_sad],
    [val_T2_rep_ang, val_T2_rep_hap, val_T2_rep_sad],
    [val_T1_zero_ang, val_T1_zero_hap, val_T1_zero_sad],
    [val_T2_zero_ang, val_T2_zero_hap, val_T2_zero_sad],
]

# Data for accuracy
acc_data = [
    [acc_T1_rep_ang, acc_T1_rep_hap, acc_T1_rep_sad],
    [acc_T2_rep_ang, acc_T2_rep_hap, acc_T2_rep_sad],
    [acc_T1_zero_ang, acc_T1_zero_hap, acc_T1_zero_sad],
    [acc_T2_zero_ang, acc_T2_zero_hap, acc_T2_zero_sad],
]


# Create subplots for validation loss
fig, ax = plt.subplots(3, 4, figsize=(20, 15))
create_subplots(ax, val_data, "Validation Loss", "Loss")

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```

```python
# Create subplots for validation loss
fig, ax = plt.subplots(3, 4, figsize=(20, 15))
create_subplots(ax, acc_data, "Accuracy", "Accuracy")

# Adjust layout and show the plots
plt.tight_layout()
plt.show()
```
