** Repeating Angry **

```python
import os
from scipy.io import wavfile
import numpy as np

# Define the source and destination directories
source_directory = "/content/drive/My Drive/DSAP/BigProject/DATA/CremaD/AudioWAV"
destination_directory = "/content/drive/My Drive/DSAP/BigProject/DATA/CremaD/AudioWAV_repeated"

# Read the files in the source directory
audio_files = os.listdir(source_directory)

# Loop through audio files and repeat audio data until it reaches the length 'n'
for audio_file in audio_files:
    # Read the audio file
    file_path = os.path.join(source_directory, audio_file)
    sampling_rate, audio_data = wavfile.read(file_path)
    # Get the length of the audio data
    len_audio = len(audio_data)
    # Desired length
    n = 80000
    if len_audio < n:
        # Calculate the number of repetitions needed
        repetitions = int(np.ceil(n / len_audio))
        # Repeat the audio data
        repeated_audio_data = np.tile(audio_data, repetitions)
        # Trim the repeated audio data to the desired length
        repeated_audio_data = repeated_audio_data[:n]
    else:
        # If the audio is longer or equal to the desired length, use the first 'n' samples
        repeated_audio_data = audio_data[:n]
    # Save the repeated audio signal as a new WAV file
    destination_path = os.path.join(destination_directory, audio_file)
    wavfile.write(destination_path, sampling_rate, repeated_audio_data.astype(np.int16))

```
